```python
import os
import pandas as pd
```

```python
weather_files = [x for x in os.listdir() if '.csv' in x]
cities = [x.split(".")[0] for x in weather_files]

weather_files
```

```python
df = pd.concat([pd.read_csv(x) for x in weather_files])

df = df.rename(columns={'name': 'city', 'datetime': 'date'})

df.head()
```

```python
keep_columns = ['city', 'date', 'tempmax', 'tempmin', 'temp', 'feelslikemax',
       'feelslikemin', 'feelslike', 'dew', 'humidity', 'precip', 'precipprob',
       'precipcover', 'snow', 'snowdepth', 'windgust', 'windspeed', 'winddir',
        'cloudcover', 'visibility', 'solarradiation', 'solarenergy',
       'uvindex', 'conditions']

df = df[keep_columns]
df.head()
```

```python
df.to_csv('weather_merged.csv', index=False)
```
