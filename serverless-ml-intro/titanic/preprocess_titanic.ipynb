```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("titanic.csv")
data.head()
```

```python
data_processed = data.copy()

# binarize sex
data_processed.Sex = data_processed.Sex.apply(lambda x: 1.0 if x == 'female' else 0.0)

# make Embarked numerical
def numerize_embarked(x):
    if x == "S":
        return 0
    elif x == "C":
        return 1
    elif x == "Q":
        return 2
    
data_processed.Embarked = data.Embarked.apply(lambda x: numerize_embarked(x))

# drop non-predictive
data_processed.drop(['PassengerId', 'Name', 'Ticket', 'Cabin'], axis=1, inplace=True)

# drop the nan
data_processed = data_processed.dropna()
```
